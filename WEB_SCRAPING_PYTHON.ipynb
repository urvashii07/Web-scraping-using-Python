# Web scrapping for television using python scrapped prices,brands and ratings data from flipkart

```python
#import all libraries for web scraping
import pandas as pd 
import numpy as np
import requests
from bs4 import BeautifulSoup
```

# scraping data using URL for APPLE Laptop

```python

laptop_name=[]
laptop_price=[]
laptop_rating=[]

url="https://www.flipkart.com/search?q=apple%20laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
req=requests.get(url)
content=BeautifulSoup(req.content,"html.parser")
#print(content)
name=content.find_all("div",{"class":"_4rR01T"})
price=content.find_all("div",{"class":"_30jeq3 _1_WHN1"})
rating=content.find_all("div",{"class":"_3LWZlK"})
#print("phone in page"+str(i))
print((name))
for i in name:
    laptop_name.append(i.text)
for i in price:
    laptop_price.append(i.text)
for i in rating:
    laptop_rating.append(i.text)
```

```python
print("laptop_name",laptop_name)
```

```python
print("laptop_price",laptop_price)
```

```python
print("laptop_rating",laptop_rating)
```

```python
print("laptop_name=",len(laptop_name))
print("laptop_price=",len(laptop_price))
print("laptop_rating=",len(laptop_rating))
```

```python
laptop_rating=laptop_rating[0:24]
print("laptop_rating=",len(laptop_rating))
```

```python
#to covert data from list into datafram
import pandas as pd
data={"Product_Name":laptop_name,"Product_Price":laptop_price,"Product_Rating":laptop_rating}
apple_laptop=pd.DataFrame(data)
apple_laptop
```
